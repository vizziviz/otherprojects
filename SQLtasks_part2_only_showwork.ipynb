```python
import sqlite3
from sqlite3 import Error


def create_connection(db_file):
    """ create a database connection to the SQLite database
        specified by the db_file
    :param db_file: database file
    :return: Connection object or None
    """
    conn = None
    try:
        conn = sqlite3.connect(db_file)
        print(sqlite3.version)
    except Error as e:
        print(e)
 
    return conn
 
def select_all_tasks(conn):
    """
    Query all rows in the tasks table
    :param conn: the Connection object
    :return:
    """
    cur = conn.cursor()
    
    query1 = """
        SELECT *
        FROM FACILITIES
        """
    cur.execute(query1)
 
    rows = cur.fetchall()
 
    for row in rows:
        print(row)


def main():
    database = "sqlite\db\pythonsqlite.db"
 
    # create a database connection
    conn = create_connection(database)
    with conn: 
        print("2. Query all tasks")
        select_all_tasks(conn)
 
```

```python
#PLAYING WITH QUERIES TO SEE THE REVENUE OF EACH BOOKING

df2 = pd.read_sql_query("""
    SELECT 
        CASE WHEN b.memid = 0 THEN (b.slots * f.guestcost) 
        ELSE (b.slots * f.membercost) END as revenue,
        f.name 
    FROM Bookings as b
    LEFT JOIN Facilities as f 
    ON b.facid = f.facid
    """, engine)
print(df2.head())
print(df2.tail())
```

```python
df4 = pd.read_sql_query("""
    SELECT
        s.name,
        SUM(s.revenue) as totalrevenue
    FROM
        (SELECT 
            CASE WHEN b.memid = 0 THEN (b.slots * f.guestcost)
            ELSE (b.slots * f.membercost) END as revenue,
            f.name as name
        FROM Bookings as b
        LEFT JOIN Facilities as f 
        ON b.facid = f.facid) as s
    GROUP BY s.name
    HAVING totalrevenue < 1000""", engine)
```

# /* Q11: Produce a report of members and who recommended them in alphabetic surname,firstname order */`

```python
recs = pd.read_sql_query("""
    SELECT m1.surname || ', ' || m1.firstname as member,
        m2.surname || ', ' || m2.firstname as recommendedby
    FROM Members as m1
    INNER JOIN Members as m2
    ON m1.recommendedby = m2.memid
    ORDER BY member
    """, engine)
```

# /* Q12: Find the facilities with their usage by member, but not guests */

```python
use = pd.read_sql_query("""
    SELECT f.name as name,
        COUNT(CASE WHEN b.memid <> 0 THEN b.bookid END) as num_bookings
    FROM Bookings as b
    INNER JOIN Facilities as f
    ON b.facid = f.facid
    GROUP BY f.name""", engine)
```

# Anyway I found another way to do it

```python
months = pd.read_sql_query("""
    SELECT f.name as name,
        strftime('%m',starttime) as Month,
        count(bookid) as num_bookings
    FROM Bookings as b
    INNER JOIN Facilities as f
    ON b.facid = f.facid
    WHERE b.memid<>0
    GROUP BY name, Month
    """, engine)
print(months)
```
